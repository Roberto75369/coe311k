```python
import numpy as np
import matplotlib.pyplot as plt
import math
import time

x_start = 0
x_end = 2

def f(x, y):
    return -2 * y

true_y = lambda x: np.exp(-2 * x)

first_run_errors = None

for h in np.arange(0.01, 1.05, 0.05):
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    y_values[0] = 1

    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])

    errors = abs(y_values - true_y(x_values))

    if first_run_errors is None:
        first_run_errors = errors.copy()

    total_error = np.sum(errors)
    print(f"h = {h:.2f}   total accumulated error = {total_error:.6f}")

    for i in range(len(x_values)):
        print(f"  x = {x_values[i]:.2f}   error vs first run = {abs(errors[i] - first_run_errors[min(i, len(first_run_errors)-1)]):.6f}")

    plt.figure()
    plt.plot(x_values, y_values, label=f"euler h = {h:.2f}")
    plt.plot(x_values, true_y(x_values), linestyle="--", label="true solution")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"approximation of dy/dx = -2y using euler method h = {h:.2f}")
    plt.legend()
    plt.grid(True)
    plt.show()
```
