```python
import numpy as np
import matplotlib.pyplot as plt
import time

#naive matrix style from the slides
def mat_mult(A, B):
    rows_A = len(A)
    cols_A = len(A[0])
    cols_B = len(B[0])
    C = [[0] * cols_B for _ in range(rows_A)]
    for i in range(rows_A):
        for j in range(cols_B):
            for k in range(cols_A):
                C[i][j] += A[i][k] * B[k][j]
    return C
    
#we will have 8 pairs each doubling in size from 3 all the way to 384

sizes = [3 * (2**i) for i in range(6)]

naive_times = []
numpy_times = []

for n in sizes:
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)

    A_list = A.tolist()
    B_list = B.tolist()

    # naive timing
    t0 = time.time()
    mat_mult(A_list, B_list)
    t1 = time.time()
    naive_times.append(t1 - t0)

    # numpy timing
    t0 = time.time()
    np.matmul(A, B)
    t1 = time.time()
    numpy_times.append(t1 - t0)

# plot
plt.plot(sizes, naive_times, label='Naive')
plt.plot(sizes, numpy_times, label='NumPy')
plt.title('Matrix Multiplication Runtime Comparison')
plt.xlabel('Matrix Size (n×n)')
plt.ylabel('Runtime (seconds)')
plt.yscale('log')
plt.legend()
plt.show()
```
