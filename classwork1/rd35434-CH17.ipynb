```python
import numpy as np 
import matplotlib.pyplot as plt
import math
import time

k = 1.0
h_values = [0.01, 0.1, 0.5, 1.0]
t_start = 0
t_end = 20
y_0 = 1.0
v_0 = 0.0
 
def euler_method(h, k, t_start, t_end, y_0, v_0):
    t_values = np.arange(t_start, t_end + h, h)
    y_values = np.zeros(len(t_values))
    v_values = np.zeros(len(t_values))
 
    y_values[0] = y_0
    v_values[0] = v_0
 
    for i in range(1, len(t_values)):
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]
 
        y_values[i] = y_n + h * v_n
        v_values[i] = v_n + h * (-k * y_n)
 
    return t_values, y_values
 
plt.figure(figsize=(12, 6))
 
for h in h_values:
    # coefficient matrix
    A = np.array([[1, h],
                  [-h * k, 1]])
 
    eigenvalues = np.linalg.eigvals(A)
    max_eigenvalue = np.max(np.abs(eigenvalues))
 
    print(f"h = {h}")
    print(f"  eigenvalues = {eigenvalues}")
    print(f"  max magnitude = {max_eigenvalue:.6f}")
 
    if max_eigenvalue < 1:
        print(f"  Solution is Stable")
    else:
        print(f"  Solution may not be Entirely Stable ;)")
    print()
 
    t, y = euler_method(h, k, t_start, t_end, y_0, v_0)
    plt.plot(t, y, label=f"Euler Method (h={h})")
 
plt.xlabel("Time (t)")
plt.ylabel("Displacement (y)")
plt.title("Stability Analysis of Euler Method for a 2nd Order ODE")
plt.legend()
plt.grid(True)
plt.show()
 
```
