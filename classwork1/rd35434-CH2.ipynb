# how to code up a way to convert 10 base numbers into 2 base numbers

```python
import numpy as np
import matplotlib.pyplot as plt

#firstly my initial thought process is to use the modulo operator and see a way to divide these numbers and letting their 
#their remainder in a if else statement to add a digit to another number

def base10_base2(n):
    if n==0:
        return '0'
#handle if the base 10 number is 0

    binary=''

    while n>0:
        r=n%2
        binary=str(r)+binary
        n=n//2

    return binary
x=base10_base2(2025)
print(x)

```
