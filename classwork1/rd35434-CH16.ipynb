```python
import numpy as np
import matplotlib.pyplot as plt 
import math
import time

x_start, x_end = 0, 10
step_sizes = [0.1, 0.5, 1, 1.5]
 
def f(x, y):
    return -2 * y
 
def exact_solution(x):
    return np.exp(-2 * x)
 
def euler_method(h):
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    y_values[0] = 1
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
    return x_values, y_values
 
plt.figure()
 
for h in step_sizes:
    x_values, y_values = euler_method(h)
    y_exact = exact_solution(x_values)
 
    local_error = np.abs(y_values - y_exact)
    global_error = np.sum(local_error)
 
    print(f"h = {h}")
    print(f"  global error = {global_error:.6f}")
    print(f"  max local error = {np.max(local_error):.6f}")
    print()
 
    plt.plot(x_values, local_error, label=f"local error h = {h}")
 
plt.xlabel("x")
plt.ylabel("error")
plt.title("local error vs x for different step sizes")
plt.legend()
plt.grid(True)
plt.show()
 
plt.figure()
 
for h in step_sizes:
    x_values, y_values = euler_method(h)
    y_exact = exact_solution(x_values)
    plt.plot(x_values, y_values, label=f"euler h = {h}")
 
x_fine = np.linspace(x_start, x_end, 1000)
plt.plot(x_fine, exact_solution(x_fine), "k--", label="exact solution")
plt.xlabel("x")
plt.ylabel("y")
plt.title("approximation of dy/dx = -2y using eulers method")
plt.legend()
plt.grid(True)
plt.show()
 
```
