```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt

sign=1
pi_divided_by_4=0

for i in range (0,5000):

    pi_divided_by_4+=sign/(2*i+1)
    sign *=-1
    pi_approx=4*pi_divided_by_4
    #print(pi_approx)
    if (i+1)%100==0:
        rel_error = abs(np.pi-pi_approx) / np.pi
        print(rel_error)
#after each term is being added, our relative error continues to decrease. the alternating sign makes the estimation increase
#and then decrease which especially rattles my brain and it makes it that much more cool how we slowly get to pi. 


print(pi_approx)
```
