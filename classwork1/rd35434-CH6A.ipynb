```python
import numpy as np
import math
import time
import matplotlib.pyplot as plt

A = np.array([[10, -1, 2, 0, 0],
              [-1, 11, -1, 3, 0],
              [2, -1, 10, -1, 0],
              [0, 3, -1, 8, -2],
              [0, 0, 0, -2, 9]],)

b = np.array([14, 30, 26, 25, 37],)

tol = 1e-6
max_iterations = 1000

def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i))
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n))
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations

x0 = np.zeros_like(b)

start = time.time()
jacobi_solution, jacobi_iterations = jacobi(A, b, x0, tol, max_iterations)
jacobi_time = time.time() - start

start = time.time()
gs_solution, gs_iterations = gauss_seidel(A, b, x0, tol, max_iterations)
gs_time = time.time() - start
#time code gathered from claud ai 
print(f"Jacobi Solution: {jacobi_solution}")
print(f"Jacobi Iterations: {jacobi_iterations}")
print(f"Jacobi Time: {jacobi_time:.6f} seconds")
print()
print(f"Gauss-Seidel Solution: {gs_solution}")
print(f"Gauss-Seidel Iterations: {gs_iterations}")
print(f"Gauss-Seidel Time: {gs_time:.6f} seconds")
#gauss seidel is slightly faster which makes sense since each time the input before is being updated in the following equation.
```
