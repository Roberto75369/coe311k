```python
import numpy as np
import matplotlib.pyplot as plt
import math
import time

a = 0
b = 2 * np.pi
N = 100
h = (b - a) / (N - 1)

x = np.linspace(a, b, N)

def f(x):
    return np.sin(x)

true_deriv = np.cos(x)

forward_diff = (f(x + h) - f(x)) / h
backward_diff = (f(x) - f(x - h)) / h
central_diff = (f(x + h) - f(x - h)) / (2 * h)

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(x, true_deriv, color="blue")
axes[0, 0].set_title("true derivative cos(x)")
axes[0, 0].set_xlabel("x")
axes[0, 0].set_ylabel("y")
axes[0, 0].grid(True)

axes[0, 1].plot(x, forward_diff, color="red")
axes[0, 1].set_title("forward difference")
axes[0, 1].set_xlabel("x")
axes[0, 1].set_ylabel("y")
axes[0, 1].grid(True)

axes[1, 0].plot(x, backward_diff, color="green")
axes[1, 0].set_title("backward difference")
axes[1, 0].set_xlabel("x")
axes[1, 0].set_ylabel("y")
axes[1, 0].grid(True)

axes[1, 1].plot(x, central_diff, color="orange")
axes[1, 1].set_title("central difference")
axes[1, 1].set_xlabel("x")
axes[1, 1].set_ylabel("y")
axes[1, 1].grid(True)

plt.suptitle("approximating derivative of f(x) = sin(x)")
plt.tight_layout()
plt.show()
```
