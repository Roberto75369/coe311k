```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(0)
violin_data = [
    np.random.normal(0, 1, 300),
    np.random.normal(2, 0.8, 300),
    np.random.normal(-1, 0.5, 300)
]
#I used ai to find out what a violin plot was and what arrays I needed to properly show it. 
#this is me getting my 3 number sets for my violin shapes, the different ranges change the size
#but the number of points in between each other is still consistent at 300


x = np.linspace(0, 2 * np.pi, 400)
xasin = np.linspace(-1, 1, 400)
tanx = np.tan(x)
tanx[np.abs(tanx) > 10] = np.nan
# i needed to generate my a new set of numbers for arcsin since the range cannot be anything and must stay within -1 to 1
#I figured out that the row and columns only allowcate the space for these graphs and only the 3rd input, the active area
#trully shows the graphs and also the order from top bottm left to right. 
plt.subplot(2, 2, 1)
plt.violinplot(violin_data, showmeans=True)
plt.xticks([1, 2, 3], ["Group A", "Group B", "Group C"])
plt.title("Violin Plot")
plt.ylabel("Value")

#the rest were trivial and a variation was close if not the same for their code in the lecture slides. 

plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x), 'r')
plt.plot(x, np.cos(x), 'g')
plt.legend(["sin(x)", "cos(x)"])
plt.title("Sine and Cosine")

plt.subplot(2, 2, 3)
plt.plot(x, tanx, 'b')
plt.title("Tangent")
plt.ylabel("y")

plt.subplot(2, 2, 4)
plt.plot(xasin, np.arcsin(xasin), 'm')
plt.title("Arcsin")
plt.tight_layout()

plt.show()
```
