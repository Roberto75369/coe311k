```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random

def samplepie(b):

    counter = 0
    for i in range(0, b):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if (x**2 + y**2 < 1):
            counter = counter + 1
    return 4 * counter / b


previous_pi = 0
true_pi = np.pi

pi_values=[]
abs_error_values=[]
real_error_values=[]
i_value=[]


for i in range(0, 100):

    calculated_pi = samplepie(1000)

    true_error = abs(calculated_pi - true_pi)
    relative_error = abs((calculated_pi - true_pi) / true_pi)

    if i > 0:
        total_calculated_pi = calculated_pi + previous_pi
    else:
        total_calculated_pi = calculated_pi

    print(calculated_pi, true_pi, relative_error, true_error)
    pi_values.append(calculated_pi)
    abs_error_values.append(true_error)
    real_error_values.append(relative_error)
    i_value.append(i)

    previous_pi = calculated_pi
#this part took a bit of help but now the difficulty is to store these values onto an array since we can't just 
#use linspace like the last time to create an array, we must manually paste these values from our for loop onto an array.
#must initialize these before the for loop. 

plt.plot(i_value,abs_error_values)
plt.xlabel("iteration")
plt.ylabel("absolute error")
plt.title("Monte Carlo absolute error")
plt.show()

plt.plot(i_value,real_error_values)
plt.xlabel("iteration")
plt.ylabel("relative error")
plt.title("Monte carlo relative error")
plt.show()

```
