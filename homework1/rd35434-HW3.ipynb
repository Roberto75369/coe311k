```python
import numpy as np
import matplotlib.pyplot as plt
import math

x_vals = [np.pi/6, np.pi/4, np.pi/3, np.pi/2]
#we must estalbish the 4 points we want to evalutae and examine.
x_labels = ["pi/6", "pi/4", "pi/3", "pi/2"]

def taylor_sin(x, num_terms):
    result = 0
    for i in range(num_terms):
        sign = (-1)**i
        numerator = x**(2*i+1)
        denominator = math.factorial(2*i+1)
        result += sign * numerator/denominator
    return result

for j in range(len(x_vals)):
    x = x_vals[j]
    print(f"x = {x_labels[j]}")
    for terms in range(1,5):
        approx = taylor_sin(x, terms)
        exact = np.sin(x)
        abs_error = abs(exact - approx)
    #not accustomed to using exact and these methods to shorten answers so used alot of influence from the slides.
    
        print(f"  {terms} term(s): approx = {approx:.6f}, exact = {exact:.6f}, abs error = {abs_error:.6e}")
    print()
    
 # I understand the use nestedfor loops to show the labels for each iteration in our second loop and in the second nested 
# to add values into our error row finishing the second loop to and then beginning our next iteration to see how these x 
# values can change as we also increase our terms through the for loop. 
# I used claude.ai to help me come in with the need for 2 nested for loops and the prompt was the 3rd through 5th bullets. 
    

x_range = np.arange(0, 10.1, 0.1)
errors = []
for terms in range(1,5):
    error_row = []
    for x in x_range:
        approx = taylor_sin(x, terms)
        exact = np.sin(x)
        abs_error = abs(exact - approx)
        error_row.append(abs_error)
    errors.append(error_row)

plt.figure()
for terms in range(1,5):
    plt.plot(x_range, errors[terms-1], label=f"{terms} term(s)")
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.title("Taylor Series Truncation Error for sin(x)")
plt.legend()
plt.yscale("log")  
plt.grid(True, alpha=0.3)  # Added gridlines to make log scale more obvious
plt.show()
#I am honestly pretty surprised at how the error is at its lowest when x is between 0-1 and one and after that it seems to 
#continue growing for the higher terms while the first term levels off. 
#as we include more terms the truncation error does decreaes when are values for x are very small. 
# It does decrease uniformly since they all mirror the same shape when x values are near or smaller than 5. 
# the error most defenitely acts different as x values increase showing that higher values of x and terms doesn't always mean 
# more accuracy truly weird to actually see. it shows how error can grow exponentially. 
```
