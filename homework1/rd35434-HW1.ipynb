```python
import matplotlib.pyplot as plt
import numpy as np
import math
import random 

def samplepie(b):

    counter = 0
    for i in range(0, b):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if (x**2 + y**2 < 1):
            counter = counter + 1
    return 4 * counter / b
sample_sizes = []
errors = []
for i in range (10,1000,20):
    calculated_error=samplepie(i)
    error=abs(calculated_error-np.pi)
    sample_sizes.append(i)
    errors.append(error)

plt.figure()
plt.plot(sample_sizes,errors)
plt.xlabel("number of samples")
plt.ylabel("absolute error")
plt.title("absolute error vs. number of samples")
plt.show()
```
