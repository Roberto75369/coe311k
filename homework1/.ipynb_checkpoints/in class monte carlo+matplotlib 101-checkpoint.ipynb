```python
import random


def samplepie(b):

    counter=0
    for i in range(0,b):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if(x**2+y**2<1):
            counter=counter+1
    return 4*counter/b
a=samplepie(5000)
print(a)
```
